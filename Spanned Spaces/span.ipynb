<h1>Part 6</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
#setting the start pos of vectors
X = np.array([0,0])
Y = np.array([0,0])
U = np.array([1,2])
V = np.array([3,-1])
#making a plot
fig, ax = plt.subplots(figsize = (5, 5))
#giving the matrix and starting pos of vectors for drawing them
plt.quiver(X, Y, U, V, angles='xy', scale_units='xy', scale=1)
 
#[-5,5] in grid
ax.set_xticks(np.arange(-5,6,1))
ax.set_yticks(np.arange(-5,6,1))
ax.axis([-5, 5, -5, 5])
#sets the lable for x and y axis
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Vectors for Matrix A")
#draw the grids
plt.grid()
#shows the plot
plt.show()
```

<h1>Part 7</h1>
<h2>The 3rd argument in arange, sets the persicion of steps :  1 =>>> -10,-9,....0,...9,10)</h2>
<h2>0.1 =>>> -10,-9.9,-9.8,......,9.9,10</h2>

```python
import numpy as np
import matplotlib.pyplot as plt
#scaler is coefficient for vectors that create space
scaler = np.arange(-5,5,0.001)
scaler = scaler.reshape(scaler.shape[0],1)
np.random.shuffle(scaler)

scaler2 = np.arange(-5,5,0.001)
scaler2 = scaler2.reshape(scaler2.shape[0],1)
np.random.shuffle(scaler2)

span_x = U[0]*scaler + U[1]*scaler2
span_y = V[0]*scaler + V[1]*scaler2


fig, ax = plt.subplots(figsize = (5, 5))
#scatter color specefic point on grid by blue dots for showing span created from 2 vectors 
#all of [scaler * x, scaler * y] are coverd by A
plt.scatter(span_x,span_y, color='b')
plt.legend()

ax.set_xticks(np.arange(-5,6,1))
ax.set_yticks(np.arange(-5,6,1))
ax.axis([-5, 5, -5, 5])
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Span created from Matrix A")
plt.grid()
plt.show()
```
